<h3> 2.5 Feature Normalization </h3>
The scale of the features (size, rooms) is not <i> very nice </i> and will cause room size to have too much weight during training. Feature normalization must be done.
<p> a feature matrix, X, can be normalized by:
$$ \frac{X - \mu_X}{\sigma_X} $$
</p>

```python
'''
In: Dataframe object, Out: Normalized version of dataframe object
'''
def normalize(df):
    if not(isinstance(df, pd.DataFrame)):
        raise Exception("Dataframe object required")
    else:
        return (df - df.mean()) / df.std()
```

```python
df_norm = normalize(df) #normalize the dataframe
```

<h3> Linear regression using BFGS </h3>
Here we are going to implement linear regression that can be used the same way as the sklearn linear regression. BFGS will be used to minimize the cost function

h(x) - Hypothesis
$$ h_\theta(X^{(i)}) - y^{(i)})  = \theta^T * X = X*\theta\ ; \theta\in R^{(n)}$$

Cost function:
$$j(\theta) = \frac{1}{2m}\sum\limits_{i=1}^m(h_\theta(x^{(i)})-y^{(i)})^2$$
$$ = \frac{1}{2m}*\sum(loss^2) $$

I will be using another dataset from the same machine learning class to debugging simple:

```python
test_data = pd.read_csv("ex1data1.csv", header=None, names=['population','profit'])
```

```python
from scipy import optimize
from scipy.optimize import minimize
```

```python
class LinearRegression():
    def __init__(self):
        self.ALPHA= .01
        self.EPSILON = 0
        self.ITER_LIMIT = 1500
        self.EPSILON = .00000001
        
    def add_bias_col(self, X):
        return np.column_stack((np.ones(X.shape[0]), X))

    def j(self, Theta, *args):
        X = args[0]
        y = args[1]
        [m, n] = X.shape
        loss = np.dot(X,Theta.reshape(n,1)) - y
        return (1/(2*m))*(sum(np.power(loss,2)))

    def pred(self, X): 
        if isinstance(X,pd.DataFrame):
            X = X.values
        X = self.add_bias_col(X)
        return np.dot(X,self.Theta.reshape(X.shape[1],1))
    
    def check_error(self, h, y):
        m = len(h)
        loss = h.reshape(m,1)-y.reshape(m,1)
        return (1/(2*m))*(sum(np.power(loss,2)))
    
    def fit(self, X, y, debug):
        #if dataframe, convert to numpy array
        
        if isinstance(X,pd.DataFrame):
            X = X.values
        if isinstance(y,pd.Series):
            y = y.values.reshape(len(y.values),1)
        
        X = self.add_bias_col(X)
        self.Theta = np.zeros((X.shape[1], 1)) #initialize Theta
        
        #minimize with BFGS using scipy optimize
        optim = minimize(self.j, self.Theta.reshape(X.shape[1],1), method='BFGS', args=(X, y), options={'disp': True if debug else False})
        self.Theta = optim['x']
```

Note: I got lazy on this part and decided not to split the data

```python
# alpha= .01
# lmbda = 0
# limit = 1500
# epsilon = .00000008
clf = LinearRegression()
clf.fit(test_data['population'],test_data['profit'], debug=True) #added a debug mode to toggle wether to display results or not
```

```python
clf.fit(X_train[['size', 'rooms']], y_train, debug=True)
```

```python
pred = clf.pred(X_test)
clf.check_error(pred, y_test.values)
```

```python
plot = plt.figure().gca(projection='3d')
plot.scatter(X_test['size'], X_test['rooms'], y_test.values)
plot.scatter(X_test['size'],X_test['rooms'],pred, color='R')
plot.set_xlabel('size')
plot.set_ylabel('rooms')
plot.set_zlabel('price')
plt.show()
```

We actually beat the cost from sklearn with this BFGS trained model at .130!

<h3> Gradient Descent </h3>

Instead of using a built in library for minimization, we will perform gradient descent to find the theta to minimize the cost function.

Gradient Descent
<p>Gradient descent is simply, the current theta - learning rate*gradient of the cost function w.r.t Theta:<br>
Here the gradient (partial derivative of the cost function wrt theta) is pre-calculated:</p>
$$\theta_j\ :=\theta_j - \alpha*\delta$$
$$\theta_j\ :=\theta_j - \alpha*\frac{1}{m}*\sum\limits_{i=1}^m((h_\theta(X^{(i)}) - y^{(i)}) * X^{(i)})$$
$$\delta = \frac{1}{m}*\sum(loss * X^{(i)}) $$
$$loss\in R^{(m)}\ X\in R^{m \times n+1}$$

```python
class LinearRegression_GD():
    def __init__(self):
        self.ALPHA= .01
        self.EPSILON = 0
        self.ITER_LIMIT = 1500
        self.EPSILON = .00000001
        
    def add_bias_col(self, X):
        return np.column_stack((np.ones(X.shape[0]), X))

    def calc_loss(self,X, y, Theta):
        n = X.shape[1]
        return np.dot(X,Theta.reshape(n,1)) - y
    
    def j(self, Theta, *args):
        X = args[0][0]
        y = args[0][1]
        [m, n] = X.shape
        loss = self.calc_loss(X,y,Theta.reshape(n,1))
        return (1/(2*m))*(sum(np.power(loss,2)))

    def pred(self, X): 
        if isinstance(X,pd.DataFrame):
            X = X.values
        X = self.add_bias_col(X)
        return np.dot(X,self.Theta.reshape(X.shape[1],1))
    
    def check_error(self, h, y):
        m = len(h)
        loss = h.reshape(m,1)-y.reshape(m,1)
        return (1/(2*m))*(sum(np.power(loss,2)))
    
    def calc_grad(self,Theta,X,y):
        [m,n] = X.shape
        loss = self.calc_loss(X,y,Theta.reshape(n,1))
        return (1/m)*np.dot(X.transpose(),loss)
        
    def gradient_descent(self, fun, x, *args, debug):
        breakpoint = -1
        old_cost = None
        history = []
        X = args[0][0]
        y = args[0][1]
        
        for i in range(1,self.ITER_LIMIT):
            cost = fun(x, (X,y))
            gradient = self.calc_grad(x,X,y)
            #break if error is tolerable
            if(i >1 and abs(cost - old_cost) <= self.EPSILON):
                old_cost = cost
                breakpoint = i
                break
            
            #update
            x = x - self.ALPHA*gradient
            old_cost = cost
            history.append(old_cost)
            
        if debug:
            print(x)
            print('final cost: ', old_cost, 'broke at:', breakpoint)
            plt.xlabel('iterations')
            plt.ylabel('cost')
            plt.plot(history)

        return {'x': x, 'cost': old_cost, 'history':history}
    
    def fit(self, X, y, debug):
        #if dataframe, convert to numpy array
        
        if isinstance(X,pd.DataFrame):
            X = X.values
        if isinstance(y,pd.Series):
            y = y.values.reshape(len(y.values),1)
        
        X = self.add_bias_col(X)
        self.Theta = np.zeros((X.shape[1], 1)) #initialize Theta
        
        #minimize with BFGS using scipy optimize
        # optim = minimize(self.j, self.Theta.reshape(X.shape[1],1), method='BFGS', args=(X, y), options={'disp': True if debug else False})
        # self.Theta = optim['x']

        #minimize using gradient descent
        optim = self.gradient_descent(self.j, self.Theta, (X,y), debug=debug)
        self.Theta = optim['x']
```

```python
# alpha= .01
# lmbda = 0
# limit = 1500
# epsilon = .00000008
clf = LinearRegression_GD()
clf.fit(test_data['population'],test_data['profit'], debug=True)
```

```python
clf.fit(X_train[['size', 'rooms']], y_train, debug=True)
```

The iterations vs cost plot is used to check if the cost is being reduced per iteration

```python
pred = clf.pred(X_test)
clf.check_error(pred, y_test.values)
```

```python
plot = plt.figure().gca(projection='3d')
print(df_norm['rooms'].shape)
plot.scatter(X_test['size'], X_test['rooms'], y_test.values)
plot.scatter(X_test['size'],X_test['rooms'],pred, color='R')
plot.set_xlabel('size')
plot.set_ylabel('rooms')
plot.set_zlabel('price')
plt.show()
```
